```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
population_df = pd.read_csv("../Resources/NST-EST2021-alldata.csv", encoding="latin-1")
population_df.head()
```

```python
database_path = ("../SQLiteDatabase/Accidents_2021.db")
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
population_df.to_sql("Population", conn, schema=None, if_exists='replace', index=False, index_label=None, chunksize=None, dtype=None, method=None)
```
